```python
X_anomalies = rng.uniform(low=-4, high=4, size=(20, 2))
```

```python
X = np.r_[X_normal, X_anomalies]
```

```python
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(X)
y_pred = model.predict(X) 
```

```python
normal_points = X[y_pred == 1]
anomaly_points = X[y_pred == -1]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(normal_points[:, 0], normal_points[:, 1], c='blue', label='Normal', alpha=0.7)
plt.scatter(anomaly_points[:, 0], anomaly_points[:, 1], c='red', label='Anomaly', marker='x', s=100)
plt.title("Isolation Forest - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
